```python
import math
import pandas as pd
```

```python
def bisection_method(a, b, e, function):
    df_method = {
        "i": [],
        "x": [],
        "f(x)": [],
        "[a, b]": [],
        "b - a": []
    }
    
    i = 1

    while abs(b-a) >= e:
        x = (a+b)/2

        f_em_a = function(a)
        f_em_x = function(x)

        if (f_em_a*f_em_x) < 0:
            b = x
        else :
            a = x

        df_method["i"].append(i)
        df_method["x"].append(x)
        df_method["f(x)"].append(f_em_x)
        df_method["[a, b]"].append("{:.7}".format(float(a)) + " ; " + "{:.7}".format(float(b)))
        df_method["b - a"].append(b-a)
        i += 1
        
    return pd.DataFrame(df_method)
```
